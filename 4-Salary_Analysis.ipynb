How well do jobs and skills pay for Data Analysts?

Methology

    1. Evaluate median salary for top 6 data jobs
    2. Find median salary per skill for Data Anlayst
    3. Visualize for highest paying skills and most demanded skills

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt
from datetime import datetime as dt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# filter for the job titles and country and drop any NaN values for 'salary_year_avg' column
df_US = df[(df['job_country'] == 'United States')].dropna(subset=['salary_year_avg'])

df_US
```

```python
# Creating a list of top 6 demanded jobs
job_titles = df_US['job_title_short'].value_counts().index[:6].tolist()

# Creating a new DataFrame with top 6 job titles
df_US_top6 = df_US[df_US['job_title_short'].isin(job_titles)]

df_US_top6
```

```python
job_title_order = df_US_top6.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
```

```python
sns.boxplot(data=df_US_top6, x='salary_year_avg', y='job_title_short', order=job_title_order)
sns.set_theme(style='ticks')

plt.title('Salary Distrubition in the United States')
plt.xlabel('Yearly Salary ($USD)')
plt.ylabel('')
plt.xlim(0, 600000)
# Formatting thee x-axis
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
plt.show()
```
